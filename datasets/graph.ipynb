```python
import numpy
import matplotlib.pyplot as plt
```

```python
gpt = [0.495, 0.0327, 0.407, 1.45]
gam = [0.622, 19.405, 9.89, 82.15]
p = [2,3,5, 9]

```

```python
plt.figure(figsize=(10, 6))
plt.plot(p, gpt, marker='o', label='SymbolicGPT')
plt.plot(p, gam, marker='o', label='GAM SymbolicGPT')

# Add title and labels
plt.xlabel('p')
plt.ylabel('MSE_N')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
